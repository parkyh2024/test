##Linear Regression란##
회귀 문제는 주어진 데이터포인트  x
 에 해당하는 실제 값으로 주어지는 타겟  y
 를 예측하는 모델
> Supervised Learning 중, 비교적 쉽고 간단한 문제

#예제#

| X | Y |
| :----: | :----: |
| 1.0 | 1.0 |
| 2.0 | 2.0 |
| 3.0 | 3.0 |
y= ax + b 의 선형 모델

x값은 예측을 위한 입력 값
y값은 입력한 값에 대한 정답값
입력값과 정답값을 기준으로 선형적인 모델을 가정하여 출발하여 학습
1차원 함수와 같이 그래프가 도출되기 때문에 아주 간단한 모델이지만 어느정도 오차가 존재한다.



##Hypothesis##

| X | Y |
| :----: | :----: |
| 1.0 | 1.0 |
| 2.0 | 2.0 |
| 3.0 | 3.0 |
y= ax + b 의 선형 모델

위와 같은 예제에서 입력(x)이 1개일 때 정답(y)이 1개인 1;1의 데이터 셋이 총 3개가 있다.  
이를 2차원에서 표시하면 3개의 점으로 표시되며 이를 1개의 선으로 이으면 1차원 함수 그래프가 나오게 된다.
이러한 1차원 함수는 선형성을 띄며 입력값이 무엇이든 1차원 함수에 대입하여 결과값을 가정할 수 있다. 
> 물론 오차가 존재하며 가정만 할 뿐

##Cost##

#조건#
Hypothesis가 정확할 수록 적은 값을, 그렇지 않을 수록 큰 값을 갖는다.

보통 mean square error를 쓴다.
$\frac{1}{N}\displaystyle\sum_{n=0}^{n=N}{(y_h-y_t)^2}$
$$
Yhypo = H(x) = w*x + b
$$

Cost가 적을 수록 좋다 = Cost를 최소화하는 w,b를 찾는다. >> 이 과정을 최적화라고 한다.



#Gradient Descent Algorithm#

![](https://static.javatpoint.com/tutorial/machine-learning/images/gradient-descent-in-machine-learning1.png)
1. 초기화(random), 초기 위치에서 탐색 시작
2. Gradient 계산
3. Gradient 방향으로 조금(Learning rate) 이동
4. 최소값으로 수렴할 때까지 반복



##Multivariable  Linear Regression ##
#Multivariable#

$$
Yhypo = H(x) = w*x + b
$$

input x가 scalar가 아닌 vector
당연히 W도 matrix가 된다.

$y_h  = \begin{bmatrix}w1&w2&w3\\ \end{bmatrix} * \begin{bmatrix}x1\\x2\\x3 \end{bmatrix}+\begin{bmatrix}b\\ \end{bmatrix}$ 

> Singlevariable에 비해 좀 더 어려운 문제지만, 문제 해결의 큰 틀은 그대로이다.
> 문제가 어려워진 경우 달라지는 점은 보통 아래와 같다.
>> 1. Data가 더 많이 필요하다.
>> 2. 연산이 더 많이 필요하다.
>> 3. Learning Rate를 줄여야 한다.
