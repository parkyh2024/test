#Tensor란#
* rank가 1인 경우 : Vector ex) 1차원 배열
* rank가 2인 경우 : Matrix ex) 2차원 배열
* rank가 2보다 큰 경우 : Tensor

Tensor 이미지

* Rank란 : 데이터를 셀 수 있는 방향의 갯수
* Shape란 : rank가 결정되었을 때 방향에 대한 데이터 갯수
* Axis란 : 데이터를 세는 방향의 순서

###데이터를 붙이는 방식###

##Concatenation##
*덩어리가 2개있을 때 경계를 허물고 붙이는 형식
##Stack##
*Axis를 지정해주면 어떤 방향으로든 붙일 수 있고 Concatenation과 달리 경계를 허물지 않는다.

###데이터를 재조합 하는 방식###

##View / Reshape##
* Data의 형태는 유지하고, rank나 axis의 길이를 조절한다.

###Tensor 연산###

##Element-wise##
* 행렬 원소별 곱셈 연산
* 연산 대상이 상수가 아니거나, Tensor의 크기가 일치하지 않는 경우, Axis의 길이와 일치해야한다.

##Transpose##
* 행렬의 행과 열을 바꾸어 놓는 연산
* ex) i,j,k -> k,j,i
* Axis 순서를 지정할 수 있으나, 잘 사용하지 않음
 
##Dot product##
* 두 벡터의 각 성분끼리의 곱의 합
* 앞 tensor의 마지막 Axis의 길이와, 뒷 tensor의 첫 Axis의 길이가 일치해야함

##Mean, Sum, Max##
* Mean : 리스트에서 최솟값을 구하는 기능
* Sum : 리스트의 값들을 더하는 기능
* Max : 리스트에서 최댓값을 구하는 기능
* Axis를 지정하여 연산 가능
