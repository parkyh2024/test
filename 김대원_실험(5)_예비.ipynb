#Multinomial Classification  
이전에 배웠던 logistic regression은 2개로 분류하는 방식인 binary classification이였지만  
* ex) binary >> 0 or 1  

multinomial classification은 3개이상으로 분류할 때 사용된다.  
* ex) multi-class claasification >> a or b or c  
![](https://media.geeksforgeeks.org/wp-content/uploads/classification-1.png)  
binary classfication에서는 sigmoid함수를 통해 0 또는 1로 구분 지었지만 multi-classfication에서는
3개이상의 구간이 필요하기 때문에 multi-level sigmoid를 사용하여 x축의 구간을 3개이상으로 만든다.
* 중간 구간과 양측의 구간간에 공평하게 데이터를 classification할 수 없음  
이를 해결하기위해 Softmax를 사용
