$$
i_L(t) = i_R(t) + i_C(t)
$$
$$
\left\{\begin{matrix}
i_R(t) = \frac{V_R}{R} = \frac{v_2(t) - v_C(t)}{R} \\ i_C(t) = -C\frac{dv_C(t)}{dt}
\end{matrix}\right.
$$  
$$
i_L(t) = \frac{1}{R}v_2(t) - \frac{1}{R}v_c(t) - C\frac{dv_C(t)}{dt}
$$
$$
\frac{dv_C(t)}{dt} = - \frac{1}{C}i_L(t) - \frac{1}{RC}v_c(t) + \frac{1}{RC}v_2(t)
$$  
**<center>[왼쪽 roop에서 KVL]</center>**   

$$
- v_1(t) + v_L(t) + v_R(t) = 0
$$
$$
v_L(t) = v_1(t) - v_R(t)
$$
$$
L\frac{di(t)}{dt} = v_1(t) - (v_2(t) - v_C(t)) = v_C(t) + v_1(t) - v_2(t)
$$
$$
\frac{di(t)}{dt} = \frac{1}{L}v_C(t) + \frac{1}{L}v_1(t) - \frac{1}{L}v_2(t)
$$  
**<center>[Differential equation]</center>**  

$$
\dot{x(t)} = \begin{vmatrix} 0 & \frac{1}{L} \\ -\frac{1}{C} & -\frac{1}{RC} \end{vmatrix} x(t) + \begin{vmatrix} \frac{1}{L} & -\frac{1}{L} \\ 0 & \frac{1}{RC} \end{vmatrix} \begin{vmatrix} v_1(t) \\ v_2(t) \end{vmatrix}

**(a) State variable model**
$$
T(s) = \frac{Y(s)}{R(s)} = \frac{8s+40}{s^3+12s^2+44s+48}
$$  
$$
Y(s) = (8s+40) \cdot Z(s)
$$
$$
R(s) = (s^3+12s^2+44s+48) \cdot Z(s)
$$   

**<center> Inverse laplace transform </center>**  

$$
y(t) = 8\dot{z(t)} + 40z(t)
$$
$$
r(t) = \dddot{z(t)} + 12\ddot{z(t)} + 44\dot{z(t)} + 48z(t)
$$  
$$
\dot{x_3}(t) = \dddot{z(t)} = - 12\ddot{z(t)} - 44\dot{z(t)} - 48z(t) + r(t)
$$  
$$
x(t) = \begin{vmatrix} x_1(t) \\ x_2(t) \\ x_3(t) \end{vmatrix} = \begin{vmatrix} z(t) \\ \dot{x_1(t)} \\ \dot{x_2(t)} \end{vmatrix} = \begin{vmatrix} z(t) \\ \dot{z(t)} \\ \ddot{z(t)} \end{vmatrix}
$$  
$$
= - 12x_3(t) - 44x_2(t) - 48x_1(t) +r(t)
$$
$$
y(t) = 8x_2(t) + 40x_1(t)
$$
$$
\dot{x(t)} = Ax(t) + Br(t)
$$  

$$
= \begin{vmatrix} 0 & 1 & 0 \\ 0 & 0 & 1 \\ -48 & -44 & -12 \end{vmatrix} x(t) + \begin{vmatrix} 0 \\ 0 \\ 1 \end{vmatrix} r(t)
$$  
$$
y(t) = \begin{vmatrix} 40 & 8 & 0 \end{vmatrix} x(t)
$$  

**(b) state transition matrix**  

**matlab을 활용한 문제풀이**

```python
syms s

% matrix 선언
A = [0, 1, 0;
     0, 0, 1;
     -48, -44, -12];
B = [0; 0; 1];
C = [40, 8, 0];
D = 0;

% state transition matrix
Phi_s = inv(s*eye(3)-A);
Phi_t = ilaplace(Phi_s, s);
pretty(Phi_t)
```

**matlab을 활용한 문제풀이**

```python
syms s

% matrix 선언
A = [1, 1, -1;
     4, 3, 0;
    -2, 1, 10
    ];
B = [0; 0; 4];
C = [1, 0, 0];
D = 0;

% state variable equation -> transfer function
Phi = inv(s*eye(3)-A) % inv = 역행렬(Matrix Inverse) // eye = 단위 행렬(Identity Matrix)
G = C * Phi * B + D
pretty(G)
```
