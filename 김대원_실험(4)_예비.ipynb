## Cost

Logistic Regression에서는 보통 Mean square error를 사용한다.

$\frac{1}{N}\displaystyle\sum_{n=0}^{n=N}{(y_h-y_t)^2}$

![](https://miro.medium.com/v2/resize:fit:970/1*_icGs0xO2Gg7-FfFVMEDdA.png)

하지만 Mean square error를 사용할 때 위 사진과 같은 문제점이 발생한다. gradient descent를 통해 문제를 풀게되면 경사를 따라 최저의 지점 찾게되는데 중간에 local cost minimum지점과 같이 안착하게되어 최저의 지점을 찾는데 어려움이 있다. 
이를 해결하기 위해 binary cross entropy가 등장한다.


$\ -[y_tlog(y_h) + (1 - y_t)log(1-y_h)]$
![](https://themaverickmeerkat.com/img/CrossEntropy/binaryCE.png)
수식을 그래프 그려보면 위와같이 그려지는데 mean square error와는 달리 local minimum과 같은 중간 지점이 생기기 않고 깔끔하게 최저 지점에 도달하는 것을 볼 수 있다. 
답이 1인데 0을 반환하게되면 loss가 무한에 가까워지고 1을 반환하면 0에 가까워지게 된다. 
이와 같이 cost가 미분가능하면서 방향성도 가지고 있으며 loss에 대한 신뢰도가 높기 때문에 많이 사용하고 있다.
